#  Python Dersi - 2. Hafta Ödevleri

Bu dosyada 2. hafta ödevleri kapsamında kullanıcıdan alınan değerlerle koşul ifadeleri, döngüler ve fonksiyonlar kullanılarak çeşitli temel Python problemleri çözülmüştür.


## 1. Girilen sayının pozitif, negatif veya sıfır olduğunu yazan koşul

```python
sayi = int(input("Bir sayı girin: "))
if sayi > 0:
    print("Pozitif sayı")
elif sayi < 0:
    print("Negatif sayı")
else:
    print("Sıfır")
```

## 2. Girilen sayının tek mi çift mi olduğunu yazan koşul

```python
sayi = int(input("Bir sayı girin: "))
if sayi % 2 == 0:
    print("Çift sayı")
else:
    print("Tek sayı")
```

## 3. Girilen nota göre harf aralığını yazan koşul

```python
notu = int(input("Notunuzu girin: "))
if 80 <= notu <= 100:
    print("A")
elif 60 <= notu < 80:
    print("B")
elif 40 <= notu < 60:
    print("C")
else:
    print("D")
```

## 4. Girilen ismin karakter sayısı 5'ten büyükse yorum

```python
isim = input("İsminizi girin: ")
if len(isim) > 5:
    print("Uzun bir isminiz var")
else:
    print("İsminiz:", isim)
```

## 5. Girilen sayının asal olup olmadığını bulan kod (for)

```python
sayi = int(input("Bir sayı girin: "))
asal = True
if sayi < 2:
    asal = False
else:
    for i in range(2, sayi):
        if sayi % i == 0:
            asal = False
            break
if asal:
    print("Asal")
else:
    print("Asal değil")
```

## 5. Girilen sayının asal olup olmadığını bulan kod (while)

```python
sayi = int(input("Bir sayı girin: "))
i = 2
asal = True
if sayi < 2:
    asal = False
else:
    while i < sayi:
        if sayi % i == 0:
            asal = False
            break
        i += 1
if asal:
    print("Asal")
else:
    print("Asal değil")
```

## 6. notlar listesinde 75'in indisini yazdıran döngü

```python
notlar = [45, 85, 75, 50]
for i in range(len(notlar)):
    if notlar[i] == 75:
        print("75'in indisi:", i)
```

## 7. Girilen sayının faktöriyeli (for)

```python
sayi = int(input("Bir sayı girin: "))
faktoriyel = 1
for i in range(1, sayi+1):
    faktoriyel *= i
print("Faktöriyel:", faktoriyel)
```

## 7. Girilen sayının faktöriyeli (while)

```python
sayi = int(input("Bir sayı girin: "))
faktoriyel = 1
i = 1
while i <= sayi:
    faktoriyel *= i
    i += 1
print("Faktöriyel:", faktoriyel)
```

## 8. Pozitif sayı girilene kadar soran yapı (while)

```python
while True:
    sayi = int(input("Pozitif bir sayı girin: "))
    if sayi > 0:
        print("Girdiğiniz pozitif sayı:", sayi)
        break
```

## 9. Fonksiyon ile asal kontrolü (for)

```python
def asal_mi(s):
    if s < 2:
        return False
    for i in range(2, s):
        if s % i == 0:
            return False
    return True

sayi = int(input("Sayı girin: "))
if asal_mi(sayi):
    print("Asal")
else:
    print("Asal değil")
```

## 9. Fonksiyon ile asal kontrolü (while)

```python
def asal_mi(s):
    if s < 2:
        return False
    i = 2
    while i < s:
        if s % i == 0:
            return False
        i += 1
    return True

sayi = int(input("Sayı girin: "))
if asal_mi(sayi):
    print("Asal")
else:
    print("Asal değil")
```

## 10. Fonksiyon ile faktöriyel (for)

```python
def faktoriyel(s):
    sonuc = 1
    for i in range(1, s+1):
        sonuc *= i
    return sonuc

sayi = int(input("Sayı girin: "))
print("Faktöriyel:", faktoriyel(sayi))
```

## 10. Fonksiyon ile faktöriyel (while)

```python
def faktoriyel(s):
    sonuc = 1
    i = 1
    while i <= s:
        sonuc *= i
        i += 1
    return sonuc

sayi = int(input("Sayı girin: "))
print("Faktöriyel:", faktoriyel(sayi))
```
